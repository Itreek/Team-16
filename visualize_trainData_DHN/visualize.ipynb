```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
file1 = open("DeepMOT-2.txt","r")
array=file1.readlines()
# print(len(array))
L1=[]
for i in range(7,len(array),4):
    L1.append((array[i]))
file1.close()
```

```python
acc=[]
for i in L1:
    # print(i)
    str=i
    for j in range(len(str)):
        if(str[j-1]=='y' and str[j]==' '):
            k=str[j+1:j+6]
            acc.append(k)
        else:
            continue
# print(len(acc))
# print(acc)
accuracy=[]
for i in acc:
    accuracy.append(float(i))
print(accuracy)
```

```python
y_value = 1
x = np.arange(10)
y = accuracy 
x=np.arange(0,1,1/594)
plt.plot(x,y, ls='dotted', c='red', lw=2)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy accross 1 epoch')
plt.show()
```

```python
prec=[]
rec=[]
for i in range(8,len(array),4):
    str=(array[i])
    for j in range(len(str)):
        if(str[j]==' ' and str[j-1]=='P'):
            k=str[j+1:j+5]
            prec.append(float(k))
        elif(str[j]==' ' and str[j-1]=='R'):
            k=str[j+1:j+5]
            rec.append(float(k))
print(len(prec))
```
